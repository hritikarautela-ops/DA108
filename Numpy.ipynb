**More useful numpy** **operation**

```python
arr = np.array([1,2,3,4])
print("Square, Squareroot, Power, Exponetial:")
print("Square:", np.square(arr))
print("Squareroot:", np.sqrt(arr))
print("Power(cube):", np.power(arr,3))
print("Exponetial:", np.exp(arr))
```

```python
#trigonometric functions
angles = np.array([0, np.pi/2, np.pi])
print("Sine:", np.sin(angles))
print("Cos:", np.cos(angles))
print("Tan:", np.tan(angles))
```

```python
#Statistical Function
data = np.array([10, 20, 30, 40, 50])
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))
print("Variance:", np.var(data))
print("Max:", np.max(data))
print("Min:", np.min(data))
```

```python
# Aggregation and cumulative operation
arr = np.array([1, 2, 3, 4])
print("sum:", np.sum(arr))
print("product:", np.prod(arr))
print("Cumulative sum:", np.cumsum(arr))
print("Cumulative product:", np.cumprod(arr))
```

**linalg**

```python
v = np.array([1,2])
print("vector:",v)
print("norm:(manhattan L1):", np.linalg.norm(v, ord=1))
print("norm (Euclidean):", np.linalg.norm(v))
print("Infinity norm:", np.linalg.norm(v,ord=np.inf))
```

```python
arr = np.array([[1,2], [3, 4]])
print("L1 norm:", np.linalg.norm(arr, ord=1))
print("L2 Norm:", np.linalg.norm(arr))
print("Infinite Norm:", np.linalg.norm(arr, ord=np.inf))
```

```python
# Matrix Multiplication
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])
print("Matrix Multiplication:\n", A @ B)
```

```python
# Determinant
arr = np.array([[1, 2],
                [3, 4]])
determinant = np.linalg.det(arr)
print("Determinant:", determinant)
```

```python
# Solving linear system Ax = b
A = np.array([[1,2],
              [3,4]])
b = np.array([5, 11])
x = np.linalg.solve(A, b)
print("Solution of Ax = b:", x)
```

```python
# Matrix Inverse
A_inv = np.linalg.inv(A)
print("Inverse of Matrix:\n", A_inv)
```

```python
# Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:", eigenvectors)
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
# Parameter for binomial distribution
n = 10
p = 0.5
size = 1000
# generate random  binomial numbers
binomial_samples = np.random.binomial(n=n, p=p, size=size)

# Print basic Information
print("Binomial Simulation:")
print("Number of Trials:", n)
print("Probability of success:", p)
print("Number of Samples:", size)
print("First !0 Samples:", binomial_samples[:10])
print("Mean of Samples:", np.mean(binomial_samples))
print("Variance of sample:", np.var(binomial_samples))
print("Standard Deviation of Samples:", np.std(binomial_samples))

#Plot Histogram
plt.figure()
plt.hist(binomial_samples, bins = range(n+2), align='left', density=True)
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution')
plt.show()
```
