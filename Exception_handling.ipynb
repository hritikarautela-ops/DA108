```python
def set_age(age):
  try:
    if age < 0:
      raise ValueError("Age cannot be negative.")
    return age
  except ValueError as e:
    print("Error:",e)

print(set_age(25))
print(set_age(-5))
```

```python
class InsufficientBalanceError(Exception):
  pass

def withdraw(amount, balance):
  if amount > balance:
    raise InsufficientBalanceError("Insufficient balance.")
    return balance - amount
try:
  withdraw(9000, 2000)
  print("Withdrawal Successful")
except InsufficientBalanceError as e:
  print("Error:",e)
```

```python
# Solution 1
balance = 9000
try:
  while True:
    amount = int(input("Enter the Withdrawal amountor (press 0 to exit): "))

    if amount == 0:
      print("Exiting...")
      break
    if amount > balance:
      raise InsufficientBalanceError("Insufficient balance.")
    else :
      balance -= amount
      print(f"Withdrawal Successful , Remaining Balance : {balance}")


except ValueError:
  print("Please , Enter a valid integer amount.")
except InsufficientBalanceError as e:
  print("Error:", e)


```

```python
# Solution 2
total = 0
count = 0
while True:
    try:
      entry = input("Enter the marks or press 'done' to quit: ")
      if entry.lower() == 'done':
        break
      marks = float(entry)
      if marks < 0 or marks > 100:
        print("Please enter marks between 0 to 100")
        continue
      total = total + marks
      count += 1

    except ValueError:
      print("Please enter a valid number.")
if count == 0:
  print("No marks entered.")
else:
  average = total/ count
  print(f"Average marks: {average:.2f}")
```
