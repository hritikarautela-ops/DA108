```python
def average_marks(rows):
  totals = {}
  counts = {}
  for x in rows :
    name, marks = x.split(",")
    if name not in totals:
      totals = 0
      counts = 0
      breakpoint()
    totals[name] += marks  #bug
    counts[name] += 1
  return {n: totals[n]/ counts[n] for n in totals}

data = ["Aman, 85",
        "Riya, 90",
        "Aman, 90",
        "Riya, 85"]

average_marks(data)

```

```python
def average_marks_fixed(rows):
  totals = {}
  counts = {}
  for x in rows :
    name, marks = x.split(",")
    name = name.strip()
    marks = int(marks.strip())

    totals[name] = totals.get(name, 0) + marks #fix
    counts[name] = counts.get(name, 0) + 1
  return {n: totals[n]/ counts[n] for n in totals}

data = ["Aman, 85",
        "Riya, 95",
        "Aman, 85",
        "Riya, 87"]

average_marks_fixed(data)
```

```python
def average_marks_robust(rows):
  totals = {}
  counts = {}

  for x in rows :
    parts = x.split(",")
    if len(parts) != 2:
      continue

    name = parts[0].strip()
    marks = parts[1].strip()

    if marks == " ":
      continue

    try:
      marks = int(marks)
    except ValueError:
      continue

    totals[name] = totals.get(name,0) + marks
    counts[name] = counts.get(name, 0) + 1
  return {n: totals[n] / counts[n] for n in totals}

average_marks_robust([
    "Aman, 80",
    "Riya, 95",
    "Aman, ",
    "Riya, NA",
    "Riya, 87"])
```

```python
def find_max(nums):
  if not nums:
    raise ValueError("Empty List")
  max_value = nums[0]
  for n in nums:
    if n > max_value:
      max_value = n
  return max_value

find_max([-5, -2, -9])
```
