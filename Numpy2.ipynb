## **Mini Project**

### Rotten Numpy Critic

```python
# Rating Movies (1.0 to 5.0)
movies = ['Avengers', 'Barbie', 'Joker', 'Intersteller', 'Parasite', 'Spiderman']
# Create a random Rating from 1.0 to 5.0 for Movies mentioned above
# Number of critics = 12
# size of Rating Grid = 12 x 6
np.random.seed(42) # Ensure rating creates same numbers for every new run of code
ratings = np.random.uniform(1.0, 5.0, size=(12, 6)).round(1)
print(ratings)
# Sum of Each Movie rating
print("Sum of Each Movie ratings:\n", ratings.sum(axis=0))
print("Average score of 'Barbie' Movie :", ratings.mean(axis=0)[1])
```

```python
# Average Rating per Movie
movie_stats = ratings.mean(axis=0)
print("Average Ratings:")
for name, score in zip(movies, movie_stats):
  print(f"{name}: {score:.2f}")

# Who is the toughest critic
critic_stats = ratings.mean(axis=1)
print("Harshest critic score:", (critic_stats.min()).round(2))
print("Most generous critic score:", (critic_stats.max()).round(2))
```

```python
# Select critic index no. 6 and add .5 rating to 6 th row
boosted_ratings = ratings.copy()
boosted_ratings[6, :] = boosted_ratings[6, :] + 0.5
# Clip to ensure noone give rating more than 5
boosted_ratings = np.clip(boosted_ratings, 1.0, 5.0)
print("Updated Ratings for critic index:6 (Critic 7):\n")
print(boosted_ratings[6, :])
```

```python
# Best review list(Rating above 4.5)
masterpiece_watch = ratings > 4.5
print("Total 5-star 'Masterpiece' reviews:", np.sum(masterpiece_watch))
print("High-rating scores:", ratings[masterpiece_watch])
```

Converting scale to 100 of first 3 rows

```python
# Vectorized multoplication
new_scale = ratings * 20
print("First 3 rows ratings converted into 100 point scale:\n", new_scale[:3])
```
