```python
# print: Is given seq. of no. is odd or even
num = int(input("Enter the number:"))
for i in range(1, num+1):
  if i % 2 == 0:
    print(f"{i} is even")
  else:
    print(f"{i} is odd")
```

```python
# Writing table using while loop
num = int(input("Enter the number:"))
i = 1

while(i < 11):
  print(f"{num} x {i} = {num*i}")
  i += 1
```

```python
# Finding factorial using for loop
def factorial(num):
  x = int(input("Enter the number:"))
  num = x
  for i in range(1, num):
    if num == 0:
      print(f"{num}!= 1")
    else:
      num = num * i
  print(f"{x}! = {num}")

factorial(5)
```

```python
# Finding factorial using while loop
def factorial(num):
  x = int(input("Enter the number:"))
  num = x
  i = num-1
  while(i > 1):
    num = num * i
    i -= 1
  print(f"{x}! = {num}")

factorial(num)
```

```python
# sum of n natural number
def sum(num):
  x = int(input("Enter the number:"))
  num = x
  sum = 0
  for i in range(1, num+1):
    sum = sum + i
  print(f"Sum of {num} natural number is {sum}")

sum(num)
```
